```python
X=[]
X.append(battingdata["'Runs'"])
X.append(battingdata["'Ave'"])
X.append(battingdata["'SR'"])
X.append(battingdata["'Fours'"])
X.append(battingdata["'Sixes'"])
X.append(battingdata["'HF'"])
X = np.asarray(X).T
#Normalize the data
X_std = StandardScaler().fit_transform(X)
#Compute the correlation matrix
cor_mat1 = np.corrcoef(X_std.T)
print('\nSample Correlation Matrix for 90 Batsmen.\n%s' %cor_mat1)
#Compute the eignevectors & eigenvalues
eig_vals, eig_vecs = np.linalg.eig(cor_mat1)

print('Eigenvectors using Cor_mat\n%s' %eig_vecs)
print('\nEigenvalues using Cor_mat\n%s' %eig_vals)

# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort(key=lambda x: x[0], reverse=True)

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
E = []
EigValEigVecPair = np.zeros((7,6))
j=0
for i in eig_pairs:
    EigValEigVecPair[1:,j] = np.asarray(i[1]).reshape(6,1).T
    E.append(i[0])
    j+=1
    
EigValEigVecPair[0,:] = np.asarray(E)
#Compute the % variability
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

#Plot the % cumulative variability
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))
    plt.bar(range(6), var_exp, alpha=0.5, align='center',
            label='individual explained variance')
    plt.step(range(6), cum_var_exp, where='mid',
             label='cumulative explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()


E = np.asarray(E)
PercentageVariability = []
for i in range(len(E)):
    PercentageVariability.append((E[i]/sum(E))*100.0)

PercentageVariability = np.asarray(PercentageVariability)
print 'Eigenvalues in descending order:\t',E
print 'Total Variability(in %):\t\t',PercentageVariability
print 'EigneValue & EigenVector Pairs for the Sample Correlation Matrix:\n',EigValEigVecPair
num_vars = 6
sing_vals = np.arange(num_vars) + 1
plt.plot(sing_vals, E, 'bo-', linewidth = 0.5),plt.title('Scree Plot for IPL 2012 Batting.')
plt.xlabel('Principal Component'),plt.ylabel('Eigenvalue')
plt.xlim([0.8,6.2]),plt.ylim([-0.5,4.5]),
plt.show()
#Function which returns the predicted ranking given the parameter values of a player X
def rankindex(obs,PC = EigValEigVecPair[1:,0]):
    return np.dot(PC,np.asarray(obs).T)
```

```python
#Compute the rank of given players and print it out
RankIdx = []
for i in range(X.shape[0]):
    RankIdx.append(rankindex(X_std[i,:]))
D = defaultdict(list)
for i in range(len(RankIdx)):
    D[RankIdx[i]].append(battingdata["'Name'"][i])
index=sorted(RankIdx)
print 'Ranking of Batsman based on the first principle component L1\n'
print 'Rank\tPlayer Name'
for i in range(len(RankIdx)):
    print i+1,'\t',D[index[len(RankIdx)-i-1]][0],'\t','{0:.2f}'.format(index[i])
```
